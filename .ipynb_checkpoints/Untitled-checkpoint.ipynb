```python
# Perceptron
import numpy as np
import random

class perceptron:
        
        def __init__(self,w=None,b=None):
            self.pesos = w		#Pesos
            self.bias  = b		#Bias
        
        


        def evalu(self,entrada):
            suma = self.bias	
            if len(self.pesos)==len(entrada):

                for i in range(len(self.pesos)):

                    suma =suma +self.pesos[i]*entrada[i]
                if suma>0:
                    print('Para la(s) entrada(s):')
                    for i in entrada:
                        print('{} '.format(i))  
                    print('El resultado es 1')

                    return 1
                else:
                    print('Para la(s) entrada(s):')
                    for i in entrada:
                        print('{} '.format(i))  
                    print('El resultado es 0')
                    return 0 
                    
            else:
                print('Entrada de tama√±o incorrecto ({}), debiera ser de largo {}'.format(len(entrada), len(self.pesos)))



per=perceptron([2,2,2,2],-7)

per.evalu([0,0,0,0]);per.evalu([1,0,0,0])
per.evalu([0,0,0,1]);per.evalu([1,0,0,1])
per.evalu([0,0,1,0]);per.evalu([1,0,1,0])
per.evalu([0,0,1,1]);per.evalu([1,0,1,1])
per.evalu([0,1,0,0]);per.evalu([1,1,0,0])
per.evalu([0,1,0,1]);per.evalu([1,1,0,1])
per.evalu([0,1,1,0]);per.evalu([1,1,1,0])
per.evalu([0,1,1,1]);per.evalu([1,1,1,1])
```
